```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
pop = pd.read_csv('C:/Users/fatem/Downloads/population.csv', skiprows=2, header=1)
pop.head()
```

## Checking whether there any Null values in Dataset

```python
pop.isna()
```

## Checking informations Regarding our Dataset like Datatype and Dimensions

```python
pop.info()
```

```python
pop.describe()
```

```python
sb.catplot(data = pop , x = 'Country Name' , y = '1960' , aspect = 4, height = 10 , kind = 'bar' )
plt.title('Country wise Population in 1960')
plt.xlabel('Country')
plt.ylabel('Year-1960')
plt.xticks(rotation=90)
plt.show()
```

```python
sb.catplot(data = pop , x = 'Country Name' , y = '2022' , aspect = 4, height = 10 , kind = 'bar' )
plt.title('Country wise Population in 2022')
plt.xlabel('Country')
plt.ylabel('Year-2022')
plt.xticks(rotation=90)
plt.show()
```
